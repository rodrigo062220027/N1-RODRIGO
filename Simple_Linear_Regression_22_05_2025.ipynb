```python
#CORREÇÕES QUESTÃO 2 - B)

import pandas as pd
from datetime import datetime

# Carregar os arquivos CSV
df_MOVIES = pd.read_csv("movies.csv")
df_RATINGS = pd.read_csv("ratings.csv")
df_TAGS = pd.read_csv("tags.csv")
df_LINKS = pd.read_csv("links.csv")

# Converter 'tmdbId' para inteiro
df_LINKS["tmdbId"] = df_LINKS["tmdbId"].fillna(0).astype(int)

# Transformar 'genres' em lista
df_MOVIES["genres"] = df_MOVIES["genres"].str.split('|')

# Converter timestamps
convert_timestamp = lambda ts: datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
df_RATINGS["timestamp"] = df_RATINGS["timestamp"].apply(convert_timestamp)
df_TAGS["timestamp"] = df_TAGS["timestamp"].apply(convert_timestamp)

# Tratar valores ausentes
df_MOVIES.fillna("Desconhecido", inplace=True)
df_TAGS.fillna("Desconhecido", inplace=True)

print("Correções aplicadas!")
print("Correções aplicadas e arquivos salvos!")
```
